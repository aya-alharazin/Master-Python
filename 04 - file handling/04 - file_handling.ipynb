# Python Files

1. What is a File? : A file is collection of bytes on disk used to store information permanently. Files allow data to persist even after a program terminates.
- **Types of Files:**
    - Text Files: Store data as human-readable text (.txt, .csv, .py, .html)
    - Binary Files: Store data in binary format (.jpg, .mp4, .exe, .pdf)
- **Why Use Files?**
    - Persistence: Data survives program termination
    - Data Sharing: Share information between programs
    - Large Data: Handle data too large for memory


2. What are File Operations?
- Open
- Read
- Write
- Close

## 3. How to Open Files

## 5. Different Ways to Read Files & Efficiency

1. Method 1: ```read()``` - Read Entire File
2. Method 2: ```readline()``` - Read One Line
3. Method 3: ```readlines()``` - Read All Lines into List
4. Method 4: Iterate Over File Object (MOST EFFICIENT!)

````python
file = open('test.txt' ,'r')
data = file.read()
print(data)
print(type(data))
file.close()
````

````python
file = open('test.txt')
line1 = file.readline()
print(line1.strip())
line2 = file.readline()
print(line2.strip())
line3 = file.readline()
print(line3.strip())
line4 = file.readline()
print(line4.strip())
file.close()
````

````python
file = open('test.txt')
list = file.readlines()
for line in list:
    print(line.strip())
file.close()
````

````python
file = open('test.txt')
for line in file:
    print(line.strip())
file.close()
````

## Encoding

````python
file = open('test5.txt' , encoding = 'utf-8')
data = file.read()
print(data)
file.close()
````

## 6. How to Write to a File
1. Method 1: write() - Write String
2. Method 2: writelines() - Write List of Strings

````python
#reading
file = open('test.txt','r')
# if test.txt exsit the file will opened for reading 
# if test.txt does not exsit FileNotFoundError
````

````python
#writing
file = open('test.txt','w')
# if test.txt exsit the file will opened for writing 
# if test.txt does not exsit it will create new file 
````

````python
file = open('out.txt','w')
file.write('aya nabil\n')
file.write('Hello world')
file.close()
````

````python
list = ['aya nabil\n' ,'saleem\n','hello world\n']
file = open('out.txt','w')
file.writelines(list)
# file.write('python is cool')
file.close()
````

## 7. How to Append to a File

````python
# Examples
file = open('out.txt','a')
file.write('\nlama , 21')
file.close()
````

## Common Pitfalls to Avoid

* ❌ Forgetting to close files
* ❌ Using 'w' when you meant 'a'
* ❌ Not handling exceptions
* ❌ Loading huge files with read()
* ❌ Forgetting to specify encoding
* ❌ Not stripping newlines when reading lines
