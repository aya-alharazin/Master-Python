## login and signup

```python
while True :
    print('1- login\n2- sign up\n3- exit')
    print('Choose 1 or 2 or 3')
    choice = input('')
    is_exsit = False
    if choice == '1' :
        entered_username = input('Enter your username')
        entered_password = input('Enter your password')
        file = open('inta.txt' , 'r')
        for line in file :
            list = line.split(',')
            username = list[0]
            password = list[1]
            if entered_username == username.strip() and entered_password == password.strip() :
                print('login successful')
                print(f"you're username is {username} " )
                print(f"you're password is {password}")
                is_exsit =True
                break
        if is_exsit == False:
            print('invalid username or password')
        file.close()
    elif choice == '2' :
        username = input('Enter a username')
        password = input('Enter a password')
        data = username+","+password+"\n"
        file = open('inta.txt' , 'a')
        file.write(data)
        file.close()
    elif choice =='3' :
        print('Good bye')
        break
    else :
        print('invalid choice')




```
