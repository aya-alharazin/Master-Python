# `File Handling`

> the process of performing operations on a file, such as creating, opening, reading, writing and closing it through a programming interface.
> It involves managing the data flow between the program and the file system on the storage device, ensuring that data is handled safely and efficiently.

## `Why do we need File Handling`
* To store data permanently, even after the program ends.
* To access external files like .txt, .csv, .json, etc.
* To process large files efficiently without using much memory.
* To automate tasks like reading configs or saving outputs.

- ###  `Opening a File`
     - Syntax:  ```file = open('filename.txt', 'mode')```

````python
# absolute path
file1 = open('C:/Users/aya/Desktop/Python for data science/Numpy/test.txt')
````

````python
# relative path
file2 = open('../test.txt')
file2.read()
````

- ### `Closing a File`
    - ```file.close()```


````python
file1.close()
````

- ### `Checking File Properties`

````python
print("Filename:", file1.name)
print("Filename:", file2.name)
print("Mode:", file1.mode)
print("Is Closed?", file1.closed)

print("Is Closed?", file2.closed)
````
